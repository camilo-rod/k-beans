```python
df = df.dropna()
```

```python
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
```

```python
df = df[
    (df['Quantity'] > 0) &
    (df['UnitPrice'] > 0)
]
```

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
reference_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)
```

```python
recency = df.groupby('CustomerID')['InvoiceDate'].max()

recency = (
    reference_date - recency
).dt.days
```

```python
frequency = df.groupby('CustomerID')['InvoiceNo'].nunique()
```

```python
monetary = df.groupby('CustomerID')['TotalPrice'].sum()
```

```python
average_purchase = monetary / frequency
```

```python
total_products = df.groupby('CustomerID')['Quantity'].sum()
```

```python
rfm = pd.DataFrame()

rfm['Recency'] = recency
rfm['Frequency'] = frequency
rfm['Monetary'] = monetary
rfm['AveragePurchase'] = average_purchase
rfm['TotalProducts'] = total_products
```

```python
rfm = rfm.replace([np.inf, -np.inf], np.nan)

rfm = rfm.dropna()
```

```python
rfm.head()
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(rfm)
```

```python
silhouette_scores = []

K = range(2,11)

best_score = -1
best_k = 2

for k in K:
    
    model = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )
    
    model.fit(X_scaled)
    
    labels = model.labels_
    
    score = silhouette_score(X_scaled, labels)
    
    silhouette_scores.append(score)
    
    print(f'Clusters: {k} | Score: {score}')
    
    if score > best_score:
        
        best_score = score
        best_k = k

print('\n=========================')
print(f'Mejor K: {best_k}')
print(f'Mejor Silhouette Score: {best_score}')
```

```python
plt.plot(K, silhouette_scores, marker='o')

plt.xlabel('Número de Clusters')
plt.ylabel('Silhouette Score')

plt.title('Selección Óptima de Clusters')

plt.show()
```

```python
kmeans = KMeans(
    n_clusters=best_k,
    random_state=42,
    n_init=10
)

kmeans.fit(X_scaled)
```

```python
rfm['Cluster'] = kmeans.labels_
```

```python
final_score = silhouette_score(
    X_scaled,
    kmeans.labels_
)

print(f'Silhouette Score Final: {final_score}')
```

```python
rfm['Cluster'].value_counts()
```

```python
cluster_analysis = rfm.groupby('Cluster').mean()

cluster_analysis
```

```python
pca = PCA(n_components=2)

X_pca = pca.fit_transform(X_scaled)
```

```python
pca_df = pd.DataFrame()

pca_df['PCA1'] = X_pca[:,0]
pca_df['PCA2'] = X_pca[:,1]

pca_df['Cluster'] = kmeans.labels_
```

```python
plt.figure(figsize=(10,7))

sns.scatterplot(
    data=pca_df,
    x='PCA1',
    y='PCA2',
    hue='Cluster',
    palette='Set2'
)

plt.title('Clusters de Clientes')

plt.show()
```

```python
cluster_names = {}

for i in range(best_k):
    
    cluster_names[i] = f'Grupo de Clientes {i}'
```

```python
rfm['Cluster_Name'] = rfm['Cluster'].map(cluster_names)
```
